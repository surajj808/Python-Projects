```python
# Perform Label Encoding for Categorical values in the data

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['gender'] = le.fit_transform(df['gender'])
df['tartar'] = le.fit_transform(df['tartar'])
df['dental caries'] = le.fit_transform(df['dental caries'])
```

# Feature Selection using Feature importance

```python
x = df.drop('smoking', axis = 1)
y = df['smoking']

from sklearn.ensemble import ExtraTreesClassifier

model = ExtraTreesClassifier()
model.fit(x,y)

df1 = pd.Series(model.feature_importances_ , index = x.columns)

plt.figure(figsize = (8,8))
df1.nlargest(24).plot(kind = 'barh')
plt.show()
```

```python
top_15 = df1.nlargest(15) # TOP 15 largest valued features selected for further.
print(top_15)
```

# Logistic Regression

```python
x = df[['gender','height(cm)','Gtp','hemoglobin','triglyceride','age','waist(cm)','HDL','weight(kg)','fasting blood sugar','LDL','ALT','systolic','serum creatinine','relaxation']]
y = df['smoking']

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=10)

x_train.shape, x_test.shape, y_train.shape
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)

x_train.shape, x_test.shape
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

lr.fit(x_train,y_train)

y_pred = lr.predict(x_test)

from sklearn.metrics import accuracy_score, classification_report

print("Accuracy: ", accuracy_score(y_test,y_pred))
print("Classification Report: \n", classification_report(y_test,y_pred))
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)

y_pred = dt.predict(x_test)
print("Classification Report: \n", classification_report(y_test,y_pred))
```

# Bagging Classifier

```python
from sklearn.ensemble import BaggingClassifier

bg = BaggingClassifier(base_estimator = DecisionTreeClassifier(), n_estimators = 1000)
bg.fit(x_train,y_train).score(x_test,y_test)

y_pred = bg.predict(x_test)
print("Classification Report: \n", classification_report(y_test,y_pred))
```

# Bagging Algorithm â€“ Extra Trees

```python
from sklearn.ensemble import ExtraTreesClassifier

et = ExtraTreesClassifier(n_estimators = 1000, random_state = 11)
et.fit(x_train,y_train)

y_pred = et.predict(x_test)
print("Classification Report: \n", classification_report(y_test,y_pred))
```

# Bagging Algorithm â€“ Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators = 1000)
rf.fit(x_train,y_train)

y_pred = rf.predict(x_test)
print("Classification Report: \n", classification_report(y_test,y_pred))
```
